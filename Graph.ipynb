```python
import matplotlib.pyplot as plt

fig = plt.figure()  # Create a figure
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # Add custom axes(Left,bottom,width,hight)
ax.plot([1, 2, 3], [4, 5, 6])  # Simple plot

plt.show()
```

```python
# prompt: create a figure with multiple subjects using 2 rows and 2 columns

import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2, ncols=2)

# Plot for subplot (0,0)
axes[0, 0].plot([1, 2, 3], [4, 5, 6])
axes[0, 0].set_title('Plot 1')

# Plot for subplot (0,1)
axes[0, 1].plot([1, 2, 3], [7, 8, 9])
axes[0, 1].set_title('Plot 2')

# Plot for subplot (1,0)
axes[1, 0].plot([1, 2, 3], [10, 11, 12])
axes[1, 0].set_title('Plot 3')

# Plot for subplot (1,1)
axes[1, 1].plot([1, 2, 3], [13, 14, 15])
axes[1, 1].set_title('Plot 4')


plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
#axes[row, col]: Access specific subplot axes.
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

# First subplot
axes[0, 0].plot([1, 2, 3], [4, 5, 6])
axes[0, 0].set_title("Plot 1")

# Second subplot
axes[0, 1].plot([1, 2, 3], [6, 5, 4])
axes[0, 1].set_title("Plot 2")

# Third subplot
axes[1, 0].plot([1, 2, 3], [1, 3, 5])
axes[1, 0].set_title("Plot 3")

# Fourth subplot
axes[1, 1].plot([1, 2, 3], [5, 3, 1])
axes[1, 1].set_title("Plot 4")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 50]
y2 = [5, 15, 20, 25, 40]

# Creating multiple subplots
plt.figure(figsize=(8, 6))

# First subplot
plt.subplot(2, 1, 1)  # (rows, columns, index)
plt.plot(x, y1, 'r-o', label='Line 1')
plt.title("First Plot")
plt.legend()

# Second subplot
plt.subplot(2, 1, 2)
plt.plot(x, y2, 'b--s', label='Line 2')
plt.title("Second Plot")
plt.legend()

# Adjust layout and show
plt.tight_layout()
plt.show()
```

```python
!pip install seaborn==0.12.2

import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for a scatter plot (e.g., relationship between two variables)
data = sns.load_dataset('iris')

# Create a scatter plot with Seaborn
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=data)
plt.title('Scatter Plot of Sepal Length vs. Sepal Width')
plt.show()

# Create a histogram with Seaborn
sns.histplot(data['sepal_length'], kde=True)
plt.title('Histogram of Sepal Length')
plt.show()

# Create a box plot with Seaborn
sns.boxplot(x='species', y='sepal_length', data=data)
plt.title('Box Plot of Sepal Length by Species')
plt.show()

# Create a violin plot with Seaborn
sns.violinplot(x='species', y='sepal_length', data=data)
plt.title('Violin Plot of Sepal Length by Species')
plt.show()

# Create a heatmap with Seaborn (e.g., correlation matrix)
# Select only numerical columns for correlation calculation
numerical_data = data.select_dtypes(include=['number'])
correlation_matrix = numerical_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Create a pair plot with Seaborn (e.g., pairwise relationships between variables)
sns.pairplot(data, hue='species')
plt.show()

# Create a joint plot with Seaborn (e.g., bivariate distribution and marginal distributions)
sns.jointplot(x='sepal_length', y='sepal_width', data=data, kind='kde')
plt.show()
```
